```python
def single_feature_predicter(pedal_length):
    if pedal_length < 2.5:
        return 0
    elif pedal_length < 4.7:
        return 1
    else:
        return 2
```

```python
manual_X = np.array([single_feature_predicter(val) for val in X_train[:, 2]])
```

```python
manual_X == Y_train
```

```python
model_accuracy = np.mean(manual_X == Y_train)
```

```python
print(f'model acvculated accuracy: {model_accuracy * 100:.2f}%')
```

```python
def plot_incorrect_preds(df_preds, x_axs, y_axs):
    fig, axs = plt.subplots(2, 2, figsize=(10, 10))
    axs = axs.flatten()
    sns.scatterplot(x=x_axs, y=y_axs, hue='prediction_name', data=df_preds, ax=axs[0])
    sns.scatterplot(x=x_axs, y=y_axs, hue='target_name', data=df_preds, ax=axs[1])
    sns.scatterplot(x=x_axs, y=y_axs, hue='correct_predictions', data=df_preds, ax=axs[2])
    axs[3].set_visible(False)

    plt.show()
```

```python
plot_incorrect_preds(df_preds, "petal length (cm)", "petal width (cm)")
```
